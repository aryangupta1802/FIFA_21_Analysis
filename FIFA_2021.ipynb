### Handling Missing Data :

```python
null_value_columns = data.loc[:,['pace',
'shooting',
'passing',
'dribbling',
'defending',
'physic',]]
```

```python
for i in null_value_columns :
    data[i].fillna(data[i].mean(),inplace=True)
```

### Analysis through Visualization

```python
%matplotlib inline
```

```python
plt.figure(figsize=(18,5))
cp = sns.countplot(x='height_cm', data = data)
cp.set_title(label = 'Players Height Distribution', fontsize = 25)
cp.set_xlabel(xlabel = 'Height', fontsize = 15)
cp.set_ylabel(ylabel = 'Player Count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(18,5))
cp = sns.countplot(x='weight_kg', data = data)
cp.set_title(label = 'Players Weight Distribution', fontsize = 25)
cp.set_xlabel(xlabel = 'Weight', fontsize = 15)
cp.set_ylabel(ylabel = 'Player Count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(18,7),dpi=125)
sns.distplot(a=data['age'],kde=False,bins=20,color='orange')
plt.axvline(x=np.mean(data['age']),c='green',ls='--',label='Mean Age of All Players')
plt.legend()
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Age Distribution', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x='height_cm', y='dribbling', data = data.sort_values('height_cm', inplace=False))
plt.xlabel('Height in cm', fontsize = 15)
plt.ylabel('Dribbling Score', fontsize = 15)
plt.title('Height v/s Dribbling', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x='weight_kg', y='dribbling', data = data.sort_values('weight_kg', inplace=False))
plt.xlabel('Weight in kg', fontsize = 15)
plt.ylabel('Dribbling Score', fontsize = 15)
plt.title('Weight v/s Dribbling', fontsize = 25)
plt.show()
```

---

```python
plt.figure(figsize=(18,5))
sns.regplot(x='weight_kg', y='pace', data = data.sort_values('weight_kg', inplace=False), color='red')
plt.xlabel('Weight in kg', fontsize = 15)
plt.ylabel('Pace', fontsize = 15)
plt.title('Weight v/s Pace', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.regplot(x='height_cm', y='pace', data = data.sort_values('height_cm', inplace=False), color='blue')
plt.xlabel('Height in cm', fontsize = 15)
plt.ylabel('Pace', fontsize = 15)
plt.title('Height v/s Pace', fontsize = 25)
plt.show()
```

```python
sns.relplot(x='overall',y='value_eur',hue='age',palette = 'magma',aspect=2,data=data)
plt.title('Overall Rating vs Value in Euros',fontsize = 25)
plt.xlabel('Overall Rating')
plt.ylabel('Value in Euros')
plt.grid()
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot('preferred_foot',data=data,palette='Reds')
plt.xlabel('Preferred Foot')
plt.ylabel('Count')
plt.title('Preferred Foot Distribution')
Right,Left=data.preferred_foot.value_counts()
print('Left Preferred',Left)
print('Right Preferred',Right)
plt.show()
```

## Goal-Keeper Analysis :

```python
gk_skills = ['gk_diving',
 'gk_handling',
 'gk_kicking',
 'gk_reflexes',
 'gk_speed',
 'gk_positioning',]
```

```python
gk_skills
```

```python
gk_data = data[data['player_positions'] == 'GK'].sort_values('overall', ascending=False)
```

```python
gk_data
```

```python
oblak = gk_data.loc[gk_data['short_name']=='J. Oblak']
oblak_df = pd.DataFrame(oblak, columns = gk_skills)

alisson = gk_data.loc[gk_data['short_name']=='Alisson']
alisson_df = pd.DataFrame(alisson, columns = gk_skills)

stegen = gk_data.loc[gk_data['short_name']=='M. ter Stegen']
stegen_df = pd.DataFrame(stegen, columns = gk_skills)

courtois = gk_data.loc[gk_data['short_name']=='T. Courtois']
courtois_df = pd.DataFrame(courtois, columns = gk_skills)

neuer = gk_data.loc[gk_data['short_name']=='M. Neuer']
neuer_df = pd.DataFrame(neuer, columns = gk_skills)
```

```python
plt.figure(figsize = (30,10))
sns.pointplot(data=oblak_df, color = 'blue')
sns.pointplot(data=alisson_df, color = 'red')
sns.pointplot(data=stegen_df, color = 'yellow')
sns.pointplot(data=courtois_df, color = 'green')
sns.pointplot(data=neuer_df, color = 'violet')

plt.title('Top 5 Goalkeeper Comparison', fontsize = 50)
plt.xlabel("skill", fontsize = 35)
plt.ylabel("skill value", fontsize = 35)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
#plt.legend(["blue", "green"], loc ="lower right")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(18,5))
cp = sns.countplot(x='height_cm', data = gk_data)
cp.set_title(label = 'Height Distribution in Goalkeepers', fontsize = 25)
cp.set_xlabel(xlabel = 'Height', fontsize = 15)
cp.set_ylabel(ylabel = 'Goalkeeper Count', fontsize = 15)
plt.show()
```
